```python
## Introducción

TourMZA es una Proof of Concept basada en Inteligencia Artificial cuyo objetivo es facilitar la planificación de visitas a bodegas en Mendoza mediante el uso de ingeniería de prompts y modelos de IA texto-texto y texto-imagen.
```

```python
## Problemática

La información sobre bodegas en Mendoza se encuentra dispersa, con diferencias en horarios, requisitos de reserva y tipos de experiencia, lo que dificulta la planificación de recorridos eficientes para turistas y visitantes.
```

```python
## Prompt Texto–Texto: Clasificación de bodegas

**Prompt:**

"Actúa como guía turístico de Mendoza. Clasifica bodegas de Luján de Cuyo según tipo de experiencia, rango de precios y horarios."

**Resultado (simulado):**

El modelo devuelve una tabla clara que permite comparar bodegas de forma rápida y estructurada.
```

```python
## Prompt Texto–Texto: Generación de itinerario

**Prompt:**

"Crea un itinerario de bodegas en Luján de Cuyo para un turista con 6 horas disponibles."

**Resultado (simulado):**

Se obtiene un recorrido optimizado con horarios, tipo de experiencia y recomendaciones de traslado.
```

```python
## Prompt Texto–Imagen

**Prompt:**

"Illustrated flat-style map of a wine tour through Luján de Cuyo, Mendoza, with vineyards and Andes mountains."

**Resultado esperado:**

Mapa ilustrado que permite visualizar el recorrido de forma clara y amigable.
```

```python
## Optimización de Prompts (Fast Prompting)

Los prompts se optimizaron definiendo roles, limitando variables y solicitando outputs estructurados, lo que permitió obtener resultados más precisos con menos iteraciones.
```

```python
## Conclusión

Esta POC demuestra que la ingeniería de prompts aplicada a modelos de IA puede mejorar significativamente la experiencia de planificación de recorridos turísticos, validando la viabilidad del proyecto TourMZA.
```
